Stationarity
In this project, the ADF will be used to check the test statistics and critical values values to determine if the dataset is stationary or not.

Stationarity can be checked using the following two methods:

Plotting Rolling Statistics: plotting the moving average or moving variance to see if it varies with time, i.e. last 12 months.

Dickey-Fuller Test: the null hypothesis is that the TS is non-stationary. In this test, results comprise of a Test Statistic and some Critical Values for difference confidence levels. If the â€˜Test Statisticâ€™ is less than the â€˜Critical Valueâ€™, the null hypothesis will be rejected and concluded that the series is stationary.

```python
# plot mean and std
plt.figure(figsize=(10,6))
rolmean = df['Price'].rolling(window=12).mean()
rolstd = df['Price'].rolling(window=12).std()
orig = plt.plot(df['Price'], color='black', label='Original')
mean = plt.plot(rolmean, color='red', label='Mean')
std = plt.plot(rolstd, color='blue', label='Std')
plt.legend(loc='best')
plt.title = ("Rolling mean and STD")
plt.show(block=False)
```

The plot shows an increasing mean. The AFD test will be used to generate the test statistics, p-value and critical values.

```python
# checking the stationarity of the data by using adfuller test
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    
test_stationarity(df['Price'])
```

According to the results of Dickey-Fuller Test, the test statistic is higher than critical values and the p-vlaue is higher than .05. This means that we failed to reject the null hypothesis and the data is non-stationary. Therefore the next step is to make the series stationary.

Estimating & Eliminating Trend
It is clearly seen that there is a significant positive trend in the dataset and transformation is necessary to penalize higher values more than smaller values. This can be resolved by taking a log, or the difference. The first step will be taking the difference and the second step will be taking a log transform to bring the data close to stationarity.

```python
# Calculating and visualizing the difference

fig, (ax1, ax2,ax3,ax4) = plt.subplots(4, 1, figsize=(10,6))
fig.suptitle('Non-Stationary into Stationary')
# Gasoline price data as a non-stationary
ax1.plot(df['Price'])

# Log Transform
ax2.plot(np.log(df['Price']))

# First Order Difference
ax3.plot(np.log(df['Price']).diff().dropna())

# Second Order Difference
ax4.plot(np.log(df['Price']).diff().diff().dropna())

ax1.title.set_text('Original Data')
ax2.title.set_text('Log Transform')
ax3.title.set_text('First Order Difference')
ax4.title.set_text('Second Order Difference')

fig.tight_layout(pad=2.5)
plt.show()
```

```python
stationary_df= np.log(df['Price']).diff().diff().dropna()
```

```python
test_stationarity(stationary_df)
```

After taking a log, or the difference, p-value has changed and not it is higher than .05.
